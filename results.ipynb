```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_blockchain = pd.read_csv('time_res_block.csv')
df_blockchain = df_blockchain.append(pd.read_csv('time_res_block1.csv'), ignore_index = True)
df_blockchain = df_blockchain.drop(['Unnamed: 0'], axis = 1)

df_blockchain = pd.read_csv('time_res_block_age.csv')
df_blockchain = df_blockchain.append(pd.read_csv('time_res_block_age1.csv'), ignore_index = True)
df_blockchain = df_blockchain.drop(['Unnamed: 0'], axis = 1)

df_blockchain.to_csv('blockchain_results.csv')
```

```python
df_blockchain
```

```python
df_blockchain.groupby(['agents']).mean().drop(['steps'], axis = 1).plot()
```

```python
df_iota = pd.read_csv('time_res_iota.csv')
df_iota = df_iota.append(pd.read_csv('time_res_iota1.csv'), ignore_index = True)
df_iota = df_iota.append(pd.read_csv('time_res_iota_age_1.csv'), ignore_index = True)
df_iota = df_iota.append(pd.read_csv('time_res_iota_age.csv'), ignore_index = True)
df_iota = df_iota.append(pd.read_csv('time_res_iota_age1.csv'), ignore_index = True)
df_iota = df_iota.append(pd.read_csv('time_res_iota_age2.csv'), ignore_index = True)
df_iota = df_iota.append(pd.read_csv('time_res_iota_sec_p.csv'), ignore_index = True)
df_iota = df_iota.append(pd.read_csv('time_res_iota_sec.csv'), ignore_index = True)
df_iota = df_iota.append(pd.read_csv('time_res_iota_sec1.csv'), ignore_index = True)

df_iota = df_iota.drop(['Unnamed: 0.1'], axis = 1)
df_iota = df_iota.drop(['Unnamed: 0.1.1'], axis = 1)
df_iota = df_iota.drop(['Unnamed: 0'], axis = 1)

df_iota['dlt'] = ['iota']*len(df_iota)

df_iota.to_csv('iota_results.csv')

df_iota
```

```python
df_iota.groupby(['agents']).mean().drop(['steps'], axis = 1).plot()
```
