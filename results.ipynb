```python
df_iota = pd.read_csv('./results/iota_results.csv')
df_iota.groupby(['agents']).mean().drop(['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.1.1', 'steps', 'Unnamed: 0.1.1.1'], axis = 1).plot()
plt.title('IOTA tiempo promedio por transacción')
plt.xlabel('Número de agentes')
plt.ylabel('Tiempo (s)')
plt.savefig('./results/iota_time_age.jpg')
```

```python
df_blockchain = pd.read_csv('./results/blockchain_results.csv')
df_blockchain.groupby(['agents']).mean().drop(['Unnamed: 0', 'Unnamed: 0.1','steps'], axis = 1).plot()
plt.title('Blockchain tiempo promedio por transacción')
plt.xlabel('Número de agentes')
plt.ylabel('Tiempo (s)')
plt.savefig('./results/block_time_age.jpg')
```
