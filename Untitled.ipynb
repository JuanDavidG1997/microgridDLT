```python
import numpy as np
import pandas as pd
```

```python
def create_synthetic_data(d_steps, d_num_agents, d_t_gens):
    demand_df = pd.DataFrame(np.random.uniform(low=30, high=35, size=(d_steps, d_num_agents)))
    price_df = pd.DataFrame(np.random.uniform(low=12, high=15, size=(d_steps, d_num_agents)))
    supply_df = pd.DataFrame(np.random.uniform(low=66, high=70, size=(d_steps, d_num_agents)))
    gens = np.random.choice(demand_df.index, size = (d_t_gens,), replace=False)
    for gen in gens:
        demand_df.iloc[gen] = np.zeros((1,d_num_agents))
        price_df.iloc[gen] = np.zeros((1,d_num_agents))
        supply_df.iloc[gen] = np.zeros((1,d_num_agents))
    return demand_df, price_df, supply_df
```

```python
demand, price, supply = create_synthetic_data(10,15, 4)
```
